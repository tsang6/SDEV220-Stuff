```python
## Thawng Sang
## File Name: gpa_qualification.py
## This app accepts student names and GPAs, then tells you wether the students qualify for the Dean's List or Honor Roll.

##The program continues to accept student entries until you type ‘ZZZ’ to exit.
DEANS_LIST_GPA = 3.5
HONOR_ROLL_GPA = 3.25

last_name = input("Enter student's last name (or 'ZZZ' to exit): ")

while last_name != 'ZZZ':
    first_name = input("Enter student's first name: ")
    gpa = float(input("Enter student's GPA: "))
    
    if gpa >= DEANS_LIST_GPA:
        print(f"{first_name} {last_name} has made the Dean's List.")
    
    if gpa >= HONOR_ROLL_GPA:
        print(f"{first_name} {last_name} has made the Honor Roll.")
    
    if gpa < HONOR_ROLL_GPA:
        print(f"{first_name} {last_name} does not qualify for either.")
    
    print()
    last_name = input("Enter student's last name (or 'ZZZ' to quit): ")

print("Program closed.")
```
