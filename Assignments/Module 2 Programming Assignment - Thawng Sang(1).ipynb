```python
secret = 3
guess = 3

if guess < secret:
    print('too low')
elif guess > secret:
    print('too high')
else:
    print('just right')
```

```python
small = True
green = False

if small and not green:
    print('cherry')
if small and green:
    print('pea')
if not small and green:
    print('watermelon')
if not small and not green:
    print('pumpkin')
```

```python
guess_me = 7
number = 1

while True:
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
    number += 1
```

```python
guess_me = 5

for number in range(10):
    if number < guess_me:
        print('too low')
    elif number == guess_me:
        print('found it!')
        break
    else:
        print('oops')
        break
```
