```python
seconds_per_day = 86400
seconds_per_hour = 3600
result = seconds_per_day / seconds_per_hour
print(result)
```

```python
seconds_per_day = 86400
seconds_per_hour = 3600
result_integer = seconds_per_day // seconds_per_hour
print(result_integer)
```

Both division methods give the same answer of 24. The only difference is that floating-point division shows 24.0 with a decimal. This makes sense because there are exactly 24 hours in a day.
