```python
from datetime import datetime
```

```python
# 13.1 
#Writes current date as a string to the text file today.txt
current_date = datetime.now()
date_as_string = current_date.strftime("%Y-%m-%d")
output_file = open('today.txt', 'w')
output_file.write(date_as_string)
output_file.close()
```

```python
# 13.2 
#Reads text file today.txt into the string `today_string`
input_file = open('today.txt', 'r')
today_string = input_file.read()
input_file.close()
```

```python
#13.3 
#Parses date from `today_string`
parsed_date = datetime.strptime(today_string, "%Y-%m-%d")

print(parsed_date)
```

```python
import multiprocessing
import time
import random
from datetime import datetime

# Function for each process
def wait_and_print():
    time.sleep(random.random())
    print(datetime.now(), flush=True)

if __name__ == '__main__':
    # Create and start three processes
    processes = []
    for i in range(3):
        process = multiprocessing.Process(target=wait_and_print)
        processes.append(process)
        process.start()

    # Wait for all processes to finish
    for process in processes:
        process.join()

# Note: Multiprocessing isn't outputting anything with Jupyter notebooks for some reason
```
