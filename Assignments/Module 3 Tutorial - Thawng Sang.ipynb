```python
def sort012(arr):
    count0 = 0
    count1 = 0
    count2 = 0
    
    for i in range(len(arr)):
        if arr[i] == 0:
            count0 += 1
        elif arr[i] == 1:
            count1 += 1
        else:
            count2 += 1
    
    i = 0
    
    # Fill 0
    while count0 > 0:
        arr[i] = 0
        i += 1
        count0 -= 1
    
    # Fill 1
    while count1 > 0:
        arr[i] = 1
        i += 1
        count1 -= 1
    
    # Fill 2
    while count2 > 0:
        arr[i] = 2
        i += 1
        count2 -= 1
    
    return arr

# Test
test_arr = [0, 1, 2, 0, 1, 2]
print(f"Before: {test_arr}")
sort012(test_arr)
print(f"After:  {test_arr}")
```

```python
def binarysearch(arr, k):
    for i in range(len(arr)):
        if arr[i] == k:
            return i
    return -1
```
